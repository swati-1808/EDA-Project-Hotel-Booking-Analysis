## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# lets check, which column has null values
missing_values[:5]

````

````python
# drop column with maximum no. of missing values
df.drop(['company'], axis=1, inplace=True)
````

````python
# As we can see there is minimal missing values in 'agent' column so replace it with 0
df['agent'].fillna(value = 0, inplace = True)
df['agent'].isnull().sum()
````

````python
# as there is less null values in 'country' so replace the null with'other' as country name
df['country'].fillna(value = 'others', inplace = True)
df['country'].isnull().sum()
````

````python
# check whether the dataset having any other null value
df.isnull().sum()
````

````python
# total stay in nights
df['total_stay_in_nights'] = df['stays_in_week_nights'] + df['stays_in_weekend_nights']
df['total_stay_in_nights']
````

````python
# Creat a new column 'revenu'
df['revenu'] = df['total_stay_in_nights'] * df['adr']
df['revenu']
````

````python
# for better understanding of 'is_canceled' col replace the value from (0,1) to 'not_canceled' , 'canceled'
df['is_canceled'] = df['is_canceled'].replace([0,1], ['not_canceled', 'canceled'])
df['is_canceled']
````

````python
# same as above for 'is_repeated_guest'
df['is_repeated_guest'] = df['is_repeated_guest'].replace([0,1] , ['not_repeated', 'repeated'])
df['is_repeated_guest']
````

````python
# Display the first few rows of the cleaned & transformed dataset
print("First few rows of the cleaned dataset:")
print(df.head())
````

#### Chart - 13

````python
# Chart - 13 visualization code
# Calculate the total number of guests for weeknights and weekend nights
total_weeknight_guests = df[df['stays_in_week_nights'] > 0]['adults'].sum() + df[df['stays_in_week_nights'] > 0]['children'].sum() + df[df['stays_in_week_nights'] > 0]['babies'].sum()
total_weekend_guests = df[df['stays_in_weekend_nights'] > 0]['adults'].sum() + df[df['stays_in_weekend_nights'] > 0]['children'].sum() + df[df['stays_in_weekend_nights'] > 0]['babies'].sum()

# Calculate the percentage of guests for weeknights and weekend nights
percentage_weeknight_guests = total_weeknight_guests / (total_weeknight_guests + total_weekend_guests) * 100
percentage_weekend_guests = total_weekend_guests / (total_weeknight_guests + total_weekend_guests) * 100

# Pie chart for the percentage of guests in weeknights and weekend nights
labels = ['Weeknights', 'Weekend Nights']
sizes = [percentage_weeknight_guests, percentage_weekend_guests]
colors = ['lightblue', 'lightcoral']

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Percentage of Guests in Weeknights vs. Weekend Nights')
plt.show()

````

##### 1. Why did you pick the specific chart?

To visualize percentage of guests in week nights and weekend nights .

##### 2. What is/are the insight(s) found from the chart?

The chart indicates that a higher percentage of guests stay during weeknights compared to weekend nights. This suggests that the hotel experiences more occupancy during weekdays. While weeknights dominate in terms of guest percentages, it also highlights an opportunity to attract more guests during weekends.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Adjusting pricing strategies, packages, and amenities based on the occupancy patterns can optimize revenue. For instance, offering special weekend packages or promotions could attract more leisure travelers, contributing to increased revenue.
